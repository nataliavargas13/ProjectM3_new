```python
import pandas as pd
import numpy as np 
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
diamonds_test = pd.read_csv('../data/diamonds_test.csv')
diamonds_test 
```

```python
cat_diamonds_test = ['cut', 'color', 'clarity', 'city']
cat_diamonds_test = diamonds_test[['cut', 'color', 'clarity', 'city']]
```

```python
#Diamonds Test
encoder_test = LabelEncoder()
cat_encod = {} 

for column in cat_diamonds_test.columns:
    cat_encod[column] = encoder_test.fit_transform(diamonds_test[column])
    
cat_encod_df = pd.DataFrame(cat_encod)
cat_encod_df
```

```python
diamonds_test = pd.concat([cat_encod_df,diamonds_test.drop(columns=['cut', 'color', 'clarity', 'city'])],axis=1)
diamonds_test
```

```python
diamonds_test.to_csv('../data/diamonds_test_enc.csv', index=False)
```

```python
diamonds_train = pd.read_csv('../data/diamonds_train.csv')
diamonds_train
```

```python
diamonds_train.dtypes
```

```python
cat_diamonds_train = ['cut', 'color', 'clarity', 'city']
cat_diamonds_train = diamonds_train[['cut', 'color', 'clarity', 'city']]
```

```python
encoder_train = LabelEncoder()
cat_encod = {} 

for column in cat_diamonds_train.columns:
    cat_encod[column] = encoder_train.fit_transform(diamonds_train[column])
    
cat_encod_df = pd.DataFrame(cat_encod)
cat_encod_df
```

```python
diamonds_train = pd.concat([cat_encod_df,diamonds_train.drop(columns=['cut', 'color', 'clarity', 'city'])],axis=1)
diamonds_train
```
